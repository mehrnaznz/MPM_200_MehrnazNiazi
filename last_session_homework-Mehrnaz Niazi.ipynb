#Database Integration Exercise

```python
df_iucn = mammals_meta.copy()

print(f"'df' DataFrame has {df.shape[0]} rows and {df.shape[1]} columns.")
print(f"'birds_meta' DataFrame has {birds_meta.shape[0]} rows and {birds_meta.shape[1]} columns.")
print(f"'df_iucn' DataFrame has {df_iucn.shape[0]} rows and {df_iucn.shape[1]} columns.")
```

```python
birds_species_only = birds_meta[birds_meta['Subsp. Seq.'] == 0.0]

print("First 5 rows of the filtered birds_species_only DataFrame:")
print(birds_species_only.head())

print(f"\nShape of birds_species_only DataFrame: {birds_species_only.shape[0]} rows, {birds_species_only.shape[1]} columns")
```

```python
unique_bird_species_count = birds_species_only['Scientific name'].nunique()
print(f"Number of unique bird species: {unique_bird_species_count}")

unique_mammal_species_count = df_iucn['Species'].nunique()
print(f"Number of unique mammal species: {unique_mammal_species_count}")
```

```python
birds_prepared = birds_species_only[['Scientific name', 'Order', 'Family name']].copy()
birds_prepared = birds_prepared.rename(columns={'Scientific name': 'Scientific Name', 'Family name': 'Family'})
birds_prepared['Class'] = 'Aves'

mammals_prepared = df_iucn[['Species', 'Order', 'Family', 'Class']].copy()
mammals_prepared = mammals_prepared.rename(columns={'Species': 'Scientific Name'})

print("Birds Prepared DataFrame (first 5 rows):")
print(birds_prepared.head())
print("\nBirds Prepared DataFrame Columns:")
print(birds_prepared.columns)

print("\nMammals Prepared DataFrame (first 5 rows):")
print(mammals_prepared.head())
print("\nMammals Prepared DataFrame Columns:")
print(mammals_prepared.columns)
```

#Perform a vertical join of Mammals and Birds data frames:

```python
metadata = pd.concat([birds_prepared, mammals_prepared], axis=0, ignore_index=True)

print("First 5 rows of the combined metadata DataFrame:")
print(metadata.head())

print(f"\nShape of metadata DataFrame: {metadata.shape[0]} rows, {metadata.shape[1]} columns")
```

#Merge the metadata dataframe with the Flavivirus dataframe:

```python
df_merged_inner = pd.merge(df, metadata, on='Scientific Name', how='inner')

print("First 5 rows of the merged DataFrame (df_merged_inner):")
print(df_merged_inner.head())

print(f"\nShape of df_merged_inner DataFrame: {df_merged_inner.shape[0]} rows, {df_merged_inner.shape[1]} columns")
```

```python
df_merged_outer = pd.merge(df, metadata, on='Scientific Name', how='outer')

print("First 5 rows of the merged DataFrame (df_merged_outer):")
print(df_merged_outer.head())

print(f"\nShape of df_merged_outer DataFrame: {df_merged_outer.shape[0]} rows, {df_merged_outer.shape[1]} columns")
```

```python
df_merged_left = pd.merge(df, metadata, on='Scientific Name', how='left')

print("First 5 rows of the merged DataFrame (df_merged_left):")
print(df_merged_left.head())

print(f"\nShape of df_merged_left DataFrame: {df_merged_left.shape[0]} rows, {df_merged_left.shape[1]} columns")
```

```python
df_merged_right = pd.merge(df, metadata, on='Scientific Name', how='right')

print("First 5 rows of the merged DataFrame (df_merged_right):")
print(df_merged_right.head())

print(f"\nShape of df_merged_right DataFrame: {df_merged_right.shape[0]} rows, {df_merged_right.shape[1]} columns")
```

```python
df_merged = df_merged_left.copy()

print("Reasoning for choosing Left Join:\n")
print("The primary goal is to combine the Flavivirus data (df) with relevant host metadata (metadata) for further analysis, specifically enriching the Flavivirus host data with taxonomic information while ensuring that ALL original Flavivirus records are retained. A left join accomplishes this by keeping all rows from the left DataFrame (df) and adding matching information from the right DataFrame (metadata). Where no match is found in the metadata for a given scientific name in df, the taxonomic columns (Order, Family, Class) are filled with NaN, indicating missing metadata rather than discarding the Flavivirus record itself.\n")

print("First 5 rows of the final merged DataFrame (df_merged):\n")
print(df_merged.head())

print(f"\nShape of df_merged DataFrame: {df_merged.shape[0]} rows, {df_merged.shape[1]} columns")
```

#Pivot table

```python
df_merged['Order'] = df_merged['Order'].fillna('Unknown Order')

pivot_table_host_diversity = pd.pivot_table(df_merged,
                                            index='Virus',
                                            columns='Order',
                                            values='Scientific Name',
                                            aggfunc='nunique')

pd.set_option('display.max_columns', None)
print("Pivot Table of Host Species Diversity by Virus and Order:\n")
print(pivot_table_host_diversity)
pd.reset_option('display.max_columns')
```
